```python
import pandas as pd
```

```python
df2017 = pd.read_html('http://www.jpx.co.jp/listing/stocks/new/index.html')[0]
df2016 = pd.read_html('http://www.jpx.co.jp/listing/stocks/new/00-archives-01.html')[0]
```

```python
def data_making(df):
    # 複数行になっているデータを一行にまとめる
    even = [x for x in range(0,len(df), 2)]
    odd = [x for x in range(1, len(df), 2)]
    
    df1 = df[df.index.isin(even)]
    df2 = df[df.index.isin(odd)]
    
    df1.index = [x for x in range(len(df1))]
    df2.index = [x for x in range(len(df2))]
    
    df1 = df1.iloc[:,[0, 1, 2, 5, 6]]
    df2 = df2.iloc[:,[0, 3, 4]]
    
    data = pd.concat([df1, df2], axis=1)
    
    data.columns = ['date', 'name', 'code', 'price_rng', 'kobo','market','price', 'uri']
    
    data = data[data.iloc[:,5]=='マザーズ']
    data.index = [x for x in range(len(data))]
    
    # ipoでないものは削除する
    delete_ind = []
    for i in range(len(data)):
        if data.iloc[i, 3] == '-':
            delete_ind.append(i)
        else:
            pass
    data = data.drop(delete_ind)
    
    data.index = [x for x in range(len(data))]
    
    return data
```

```python
data2017 = data_making(df2017)
data2016 = data_making(df2016)
```

```python
def bunKats(df):
    #　一つのセルに複数の情報が入っている場合、1セル１上方にする。
    #  いらないデータは削除する
    
    # 日付を二つに分割
    
    dated = {}
    for i in range(len(df)):
        d1 = df.iloc[i, 0].replace('）', '')
        d2 = d1.split('（')
        dated[i] = d2
        
    dated = pd.DataFrame(dated).T
    dated.columns = ['ipo_date', 'ano_date']
    
    # 社名のところの代表者インタビューのところを削除
    Namet = {}
    for i in range(len(df)):
        d1 = df.iloc[i, 1].replace('代表者インタビュー', '')
        Namet[i] = d1
        
    Namet = pd.Series(Namet)
    
    # 価格レンジを二つに分割
    Prices = {}
    for i in range(len(df)):
        d1 = df.iloc[i, 3].replace(',', '')
        d2 = d1.split('～')
        Prices[i] = d2
        
    Prices = pd.DataFrame(Prices).T
    Prices.columns = ['price_l', 'price_h']
    
    # オーバーアロットメントも分割
    #　OAを抜こうとしたけど出来ず。
#    alot = {}
#    for i in range(len(df)):
#        d1 = df.iloc[i, 7].replace(')', '')
#        d2 = d1.split('(OA')
#        alot[i] = d2
        
#    alot = pd.DataFrame(alot).T
#    alot.columns = ['urida', 'alotment']
    
    df1 = pd.concat([df, dated, Namet, Prices], axis=1)
    
    df1 = df1.loc[:, ['ano_date', 'ipo_date', 0, 'code', 'price_l', 'price_h', 'price', 'market', 'kobo', 'uri']]
    df1.columns = ['ano_date', 'ipo_date', 'name', 'code', 'price_l', 'price_h', 'price', 'market', 'kobo', 'uri']
    return df1
```

```python
data20171 = bunKats(data2017)
data20161 = bunKats(data2016)
```

```python
def make_para(df):
    data = {}
    data = pd.DataFrame(data)
    
    for i in range(0, len(df.columns)):
        d1 = df.iloc[:, i].dropna()
        if len(d1) >= 100:
            data[i] = d1[1:101].values
        else:
            pass
    return data
```
